### Import packages

```python
from transformers import pipeline
```

### We define a summarization task

```python
from transformers import pipeline

# create pieline for QA
qa = pipeline('question-answering')
```

### We do an initial inference

```python
ctx = "My name is Bob and I am studying Data Science"
que = "What is Bob studying?"
qa(context = ctx, question = que)
```
